```python
df=df.drop('Unnamed: 0', axis=1)
df.head()
```

```python
df.describe()
```

Replacing null values with mean(integers also can use median) and mode(objects-strings)

```python
#EthnicGroup
mEg=df['EthnicGroup'].mode()[0]
df['EthnicGroup']=df['EthnicGroup'].fillna(mEg)
#ParentEduc
mPe=df['ParentEduc'].mode()[0]
df['ParentEduc']=df['ParentEduc'].fillna(mPe)
#TestPrep
mTp=df['TestPrep'].mode()[0]
df['TestPrep']=df['TestPrep'].fillna(mTp)
#ParentMaritalStatus
mPms=df['ParentMaritalStatus'].mode()[0]
df['ParentMaritalStatus']=df['ParentMaritalStatus'].fillna(mPms)
#PracticeSport
mPs=df['PracticeSport'].mode()[0]
df['PracticeSport']=df['PracticeSport'].fillna(mPs)
#IsFirstChild
mC=df['IsFirstChild'].mode()[0]
df['IsFirstChild']=df['IsFirstChild'].fillna(mC)
#NrSiblings
mS=df['NrSiblings'].mean()
df['NrSiblings']=df['NrSiblings'].fillna(mS)
#TransportMeans
mTm=df['TransportMeans'].mode()[0]
df['TransportMeans']=df['TransportMeans'].fillna(mTm)
#WklyStudyHours
mSh=df['WklyStudyHours'].mode()[0]
df['WklyStudyHours']=df['WklyStudyHours'].fillna(mSh)
df.isnull().sum()
```

```python
df['TotalScore']=df['MathScore']+df['ReadingScore']+df['WritingScore']
df.head()
```

```python
#count of female and male
plt.figure(figsize=(3,5))
ax = sns.countplot(data=df, x='Gender')
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
#The graph suggests that the number of females in the dataset is higher than the number of males.
```

```python
gb = df.groupby('ParentEduc').agg({'MathScore': 'mean',	'ReadingScore': 'mean', 'WritingScore': 'mean'})
print(gb)
```

```python
plt.figure(figsize=(5,3))
sns.heatmap(gb, annot=True)
plt.title("Parent's Education vs Student's score")
plt.show()
#From this graph, we can conclude that students whose parents hold a master's degree tend to have higher scores
#while those whose parents have only completed some high school tend to have lower scores.
```
